# Exploratory Analysis

```python
# imports 
import pandas as pd
import numpy as np
```

## Loading the Data

```python
target_df = "wildfire_sizes_before_2010.csv"  # "/kaggle/input/forest-fire-prediction-epoch-hackathon/wildfire_sizes_before_2010.csv"
target_df = pd.read_csv(target_df)
target_df = target_df.rename(columns={'STATE': 'State', 'month': 'year_month'})
target_df.head()
```

```python
states_df = "merged_state_data.csv"  # "/kaggle/input/forest-fire-prediction-epoch-hackathon/merged_state_data.csv"
states_df = pd.read_csv(states_df)
states_df.head()
```

```python
weather_df = "weather_monthly_state_aggregates.csv"  # "/kaggle/input/forest-fire-prediction-epoch-hackathon/weather_monthly_state_aggregates.csv"
weather_df = pd.read_csv(weather_df)
weather_df.head()
```

```python
states_df.columns
```

```python
len(states_df)
```

```python
combined_df = pd.merge(weather_df, states_df, on='State', how='inner')
combined_df_clean = pd.merge(combined_df, target_df, on=['State', 'year_month'], how='inner')
combined_df_clean.head()
```

```python
combined_df_all = pd.merge(combined_df, target_df, on=['State', 'year_month'], how='left')
print("# rows with NaNs: ", len(combined_df_all) - len(combined_df_clean))
print("# total clean rows: ", len(combined_df_clean))
combined_df_all.head()
```
